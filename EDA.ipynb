```python
eda.ipynb

import pandas as pd
import matplotlib.pyplot as plt

import pandas as pd

df = pd.read_csv('data/clean/merged_stringency_unemployment.csv')
print(df.head())
# Load dataset
df = pd.read_csv('../data/clean/merged_stringency_unemployment.csv')

# View basic info
print(df.head())
print(df.info())
print(df.describe())

#The dataset contains merged information on 
#unemployment rates and lockdown stringency for 
#Victoria and New South Wales. Initial inspection 
#shows that the dataset includes time-series observations 
#with no major missing values after the data cleaning process.
# The variables appear consistent with expectations and suitable 
# for further analysis.

print(df.columns)


# =========================
# IMPORT LIBRARIES
# =========================
import pandas as pd
import matplotlib.pyplot as plt
# =========================
# LOAD DATA
# =========================
# Adjust path if needed later
df = pd.read_csv('data/clean/merged_stringency_unemployment.csv')

# =========================
# BASIC DATA INSPECTION
# =========================
print("First 5 rows:")
print(df.head())

print("\nColumn names:")
print(df.columns)

print("\nDataset info:")
print(df.info())

print("\nSummary statistics:")
print(df.describe())

# =========================
# DATE FORMATTING
# =========================
# Convert Date column to datetime (adjust column name if needed)
df['Date'] = pd.to_datetime(df['Date'])

# Sort by date
df = df.sort_values('Date')

# =========================
# TIME SERIES: UNEMPLOYMENT
# =========================
plt.figure(figsize=(10,5))

plt.plot(df['Date'], df['VIC_unemployment'], label='Victoria')
plt.plot(df['Date'], df['NSW_unemployment'], label='NSW')

plt.title('Unemployment Rates Over Time')
plt.xlabel('Date')
plt.ylabel('Unemployment Rate')
plt.legend()
plt.xticks(rotation=45)

plt.show()

# =========================
# TIME SERIES: STRINGENCY
# =========================
plt.figure(figsize=(10,5))

plt.plot(df['Date'], df['VIC_stringency'], label='Victoria')
plt.plot(df['Date'], df['NSW_stringency'], label='NSW')

plt.title('Lockdown Stringency Over Time')
plt.xlabel('Date')
plt.ylabel('Stringency Index')
plt.legend()
plt.xticks(rotation=45)

plt.show()

# =========================
# CORRELATION ANALYSIS
# =========================
corr_matrix = df[['VIC_unemployment', 'VIC_stringency',
                  'NSW_unemployment', 'NSW_stringency']].corr()

print("\nCorrelation Matrix:")
print(corr_matrix)

# =========================
# SCATTER PLOT (RELATIONSHIP)
# =========================
plt.figure(figsize=(6,4))

plt.scatter(df['VIC_stringency'], df['VIC_unemployment'])

plt.title('Victoria: Stringency vs Unemployment')
plt.xlabel('Stringency Index')
plt.ylabel('Unemployment Rate')

plt.show()

# Difference between states
df['unemployment_gap'] = df['VIC_unemployment'] - df['NSW_unemployment']

plt.figure(figsize=(10,5))
plt.plot(df['Date'], df['unemployment_gap'])
plt.title('Unemployment Gap (VIC - NSW)')
plt.xlabel('Date')
plt.ylabel('Difference')
plt.xticks(rotation=45)
plt.show()
#The unemployment gap between Victoria and New South Wales widens during the COVID-19 period, suggesting a stronger labour market impact in Victoria. This provides preliminary evidence supporting the treatment effect of stricter lockdowns.


```
